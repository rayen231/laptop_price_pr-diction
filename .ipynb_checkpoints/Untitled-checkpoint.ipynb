# Preprocessing

```python
#remove the uncessary columns ( bcz Product can not be encoded with that variety of types)
df=df.drop("Product",axis=1)
```

```python
#removing KG and GB
df['Ram']=df['Ram'].str.replace('GB',"").astype(int)
df['Weight']=df['Weight'].str.replace('kg',"").astype(float)
df
```

```python
#now lets encode the company column : as we say in exploration this column can be encoded but her encoding will be different 
#we are not gonna encode them like 1 for apple and 3 for hp eg bcz thats will have bad impact on our model bcz model can read 
#those values and think that they are continous and those values will lose value if u see display ...
df=df.join(pd.get_dummies(df.Company))
df.drop("Company",axis=1,inplace=True)
df
```

```python
df=df.join(pd.get_dummies(df.TypeName))
df=df.drop("TypeName",axis=1)
df
```

```python
df["ScreenResolution"]=df.ScreenResolution.str.split(" ").apply(lambda x:x[-1])
df['heigh']=df.ScreenResolution.str.split("x").apply(lambda x:x[0]).astype(int)
df['width']=df.ScreenResolution.str.split("x").apply(lambda x:x[1]).astype(int)
df.drop("ScreenResolution",axis=1,inplace=True)
```

```python
df
```

```python
df["Cpu Brand"]=df.Cpu.str.split(" ").apply(lambda x:x[0])
df["Cpu freq"]=df.Cpu.str.split(" ").apply(lambda x:x[-1])
df.drop("Cpu",axis=1,inplace=True)
```

```python
df
```

```python
df["Cpu freq"]=df["Cpu freq"].str[:-3].astype(float)#removing the unit
```

```python
df["memory_storage"]=df.Memory.str.split(" ").apply(lambda x:x[0])
df["memory_Type"]=df.Memory.str.split(" ").apply(lambda x:x[1])
```

```python
def clean(value):
    if "GB" in value:
        return float(value[:value.find("GB")])
    if "TB" in value:
        return float(value[:value.find("TB")])*1024
```

```python
df["memory_storage"]=df["memory_storage"].apply(clean)
df.drop("Memory",axis=1,inplace=True)
```

```python
df.join(pd.get_dummies(df.OpSys))
df.drop("OpSys",axis=1,inplace=True)
```

```python
df
```

```python
df["Gpu"]=df.Gpu.str.split(" ").apply(lambda x:x[0])
```

```python
df
```

```python
df=df.join(pd.get_dummies(df.memory_Type))
```

```python
cpus=pd.get_dummies(df["Cpu Brand"])
cpus.columns=[col + "_CPU" for col in cpus]
```

```python
df=df.join(cpus)
```

```python
df=df.drop("Cpu Brand",axis=1)
```

```python
Gpus=pd.get_dummies(df["Gpu"])
Gpus.columns=[col + "_GPU" for col in Gpus]
df=df.join(Gpus)
df=df.drop("Gpu",axis=1)
```

```python
df=df.drop("memory_Type",axis=1 )
```

```python
df.corr()["Price_euros"].plot(kind="bar",width=0.8)#reading the corr to see what effects the price the most 
```

```python
label=df["Price_euros"]#Y
features=df.drop("Price_euros",axis=1)#X
# --> aX+b=Y
#scaler=StandardScaler()
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,label,test_size=0.2,random_state=42,shuffle=True)
#x_train_scaled=scaler.fit_transform(x_train)
#x_test_scaled=scaler.transform(x_test)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score,accuracy_score
import numpy as np

# Convert DataFrame to NumPy array
y_test_array = y_test.to_numpy()

# Ensure predictions array has the same number of samples as y_test
predictions_array = predictions.reshape(-1, 1)

# Calculate mean squared error (MSE)
mse = mean_squared_error(y_test_array, predictions_array)

# Calculate root mean squared error (RMSE)
rmse = np.sqrt(mse)

# Calculate mean absolute error (MAE)
mae = mean_absolute_error(y_test_array, predictions_array)

# Calculate R-squared (RÂ²)
r_squared = r2_score(y_test_array, predictions_array)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (RÂ²):", r_squared)

```

```python
plt.scatter(y_test, predictions, color='blue', label='Actual vs Predicted')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2, label='Ideal line')
plt.title('Actual vs Predicted')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.legend()
plt.grid(True)
plt.show()
```
